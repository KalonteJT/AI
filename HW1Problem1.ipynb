```
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-white')
from mpl_toolkits import mplot3d
```

```
figure = plt.figure()
ax = plt.axes(projection = '3d')
def func(x,y):
  return -(x ** 2) - (y**2)

x = np.linspace(-5,5,30)
y = np.linspace(-5,5,30)

X, Y = np.meshgrid(x,y)
Z = func(X,Y)

ax.contour3D(X, Y, Z, 50, cmap='PiYG')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.plot([0],[0],[0],markerfacecolor='g', markeredgecolor = 'b', marker = 'o', markersize = 10, zorder = 100)
```
