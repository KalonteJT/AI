# ***General***

**What is Artificial Intelligence?**
The development of computer systems to perform tasks normally that require human intelligence. Consists of various techniques that allow machines to mimic human behavior.

**What is Machine Learning?**
Machine learning is a subset of Artificial Intelligence. Using machine learning a computer system can learn and improve from experience without being explicitly programmed through the use of statistical models/methods. It's goal is to allow computers to learn automatically without human intervention or assistance.
Two machine learning methods we discussed in this class are supervised learning and unsupervised learning.

**Supervised** **Learning** is the process of learning a function by mapping an input to an output based on provided input/output pairs.

**Unsupervised Learning** is a type of machine learning that allows you to draw inferences from datasets that aren't explicitly labeled.

**What is Deep Learning?**
Also known as deep neural learning, deep learning is a subset of Machine learning that allows computers to mimic the workings of a human brain when it comes to processing data and creating patterns. Deep learning algorithms use neural networks, a set of algorithms modeled after a human brain, to recognize patterns.

# ***Building a Model***

**Linear Regression Models**
An approach to modeling the relationship between a scalar dependent variable y to one or more independents variable x (ex: x1, x2, x3, ...)
Follows the equation **y = b + w1 X x1** where y is the prediction, b is the model bias, x1 is the independent variable, and w1 is the weight assigned to this prediction. 

**ConvNets**

Stands for Convolution Neural Networks and are a subcategory of neural networks that excel in areas such as image recoginition and classification.

**Structure of a ConvNet**

ConvNets consists of the following components:

*   **Convolution Layer** - The layer responsible for carrying out the convolution operation
*   **Pooling Layer** - Responsible for reducing the spatial size of the convoluted layer in a effort to decrease computational power required to process the layers data
*   **Fully Connected Layer** - The end of the neural network. Responsible for taking all the input volume and outputs an N dimension vector where N is the number of classifications the program has to choose from










# ***Compiling a model***

*    **Loss** **function** - A method for evaluating how well the model's output data represents the training data used to create the ConvNet model. Focus is to describe how well the model performed for the dataset. High loss indicates the model does not represent the data well and vice versa for low loss.
*    **Optimizer** - Uses the loss function described above to tweak the neural network to minimize loss. Focuses on network evolution.
*    **Learning Rate** - A hyperparameter that controls how much we need to adjust our CNN with respect to the loss gradient



# ***Training a Model***

**Overfitting**

Typically occurs when the model is too complex in its representation of the data. Is when the model exhibits low loss on training datasets but performs poorly on testing dataset. This is due to the model creating rules that work well for the training dataset but are not general enough to apply to new dataset to come.

**Underfitting**

Typically occurs when a model is too simple to represent the data. Performs poorly when exposed to new datasets due to the model not being able to create adequate rules to interpret the data. Exhibits high loss.

# ***Finetuning a Model***

There are many ways we can finetune a model to avoid overfitting and underfitting. One method is truncate the last layer of a pre-trained network and replace it with a softmax layer that is relevant to the problem. Another way is to use a smaller learning rate as compared to last time. Another way you could fine tune a model is to freeze the weights of the first few layers on a pre-trained CNN.
