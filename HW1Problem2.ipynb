```
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```
np.random.seed(50)
scatterX1 = 2 * np.random.rand(100,1)
scatterX2 = 2 * np.random.rand(100,1)
scatterY = 5 + 7 * scatterX1 + 3 * scatterX2 + np.random.rand(100,1)

x = scatterX1.flatten()
y = scatterX2.flatten()
z = scatterY.flatten()

figure = plt.figure(figsize=(10,10))
ax = figure.add_subplot(111,projection='3d')

ax.scatter3D(x,y,z, c = 'blue', marker = 'o')
plt.xticks(np.arange(0,2,0.5))
plt.yticks(np.arange(0,2,0.5))
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')

plt.show()


```

```
training_set1 = scatterX1[:80]
training_set2 = scatterX2[:80]
training_setY = scatterY[:80]

test_set1 = scatterX1[80:]
test_set2 = scatterX2[80:]
test_setY = scatterY[80:]
```

```
epochs = 10
lr = 0.01

w1 = np.random.randn(1)
w2 = np.random.randn(1)
b = np.zeros(1)


for epoch in np.arange(epochs):
  for i in np.arange(80):
    y_pred = w1 * training_set1[i] * training_set2[i] + b
    grad_w1 = (y_pred-training_setY[i]) * training_set1[i]
    grad_w2 = (y_pred-training_setY[i]) * training_set2[i]
    
    grad_b  = (y_pred - training_setY[i])
    
    w1 -= lr * grad_w1
    w1 -= lr * grad_w2
    b -= lr * grad_b
```

```
test_loss = 0
for i in np.arange(20):
  test_loss += 0.5 * (w1 * test_set1[i] + w2 * test_set2[i] + b - test_setY[i])**2
  test_loss /= 20
```

```
print(test_loss)
```

```
prediction_Ys = w1 * test_set1 + w2 * test_set2 +b

figure = plt.figure(figsize = (10,10))
ax = figure.add_subplot(111,projection='3d')

ax.scatter3D(test_set1, test_set2, test_setY, c = 'blue', marker = 'o')
ax.scatter3D(test_set1, test_set2, prediction_Ys, c = 'purple', marker = 'o')
plt.xticks(np.arange(0,2,0.5))
plt.yticks(np.arange(0,2,0.5))

ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')
plt.show()
```
